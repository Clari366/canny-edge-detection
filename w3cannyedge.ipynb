```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('butterfly.png', 0)  # 0 for grayscale
```

```python
edges1 = cv2.Canny(img, 50, 150)      # Default settings
edges2 = cv2.Canny(img, 30, 100)      # Lower thresholds (more edges)
edges3 = cv2.Canny(img, 100, 200)     # Higher thresholds (fewer edges)
edges4 = cv2.Canny(img, 50, 150, apertureSize=3)  # Different aperture
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 2)
plt.imshow(edges1, cmap='gray')
plt.title('Canny (50, 150) - Default')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 3)
plt.imshow(edges2, cmap='gray')
plt.title('Canny (30, 100) - Lower thresholds')
plt.axis('off')

```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 4)
plt.imshow(edges3, cmap='gray')
plt.title('Canny (100, 200) - Higher thresholds')
plt.axis('off')
```

```python
def interactive_canny():
    def nothing(x):
        pass
    
    cv2.namedWindow('Canny Edge Detection')
    cv2.createTrackbar('Low Threshold', 'Canny Edge Detection', 50, 255, nothing)
    cv2.createTrackbar('High Threshold', 'Canny Edge Detection', 150, 255, nothing)
    
    while True:
        low = cv2.getTrackbarPos('Low Threshold', 'Canny Edge Detection')
        high = cv2.getTrackbarPos('High Threshold', 'Canny Edge Detection')
        
        edges = cv2.Canny(img, low, high)
        
        cv2.imshow('Original', img)
        cv2.imshow('Canny Edge Detection', edges)
        
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            break
    
    cv2.destroyAllWindows()
```

```python
plt.tight_layout()
plt.show()

# Save results
cv2.imwrite('edges_default.jpg', edges1)
cv2.imwrite('edges_low.jpg', edges2)
cv2.imwrite('edges_high.jpg', edges3)
```
